# Sentence TF

```python
!pip install sentence_transformers --quiet
```

```python
import torch
from sentence_transformers import SentenceTransformer, util

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2', device=device)
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from itertools import product
from sklearn.decomposition import PCA

embeddings_dict = {}
pca_embeddings_dict = {}
combinations = list(product(zillow_df['city'].unique(), [0, 1])) # city, single_family
zillow_subsets = {combination: zillow_df[(zillow_df['city'] == combination[0]) & (zillow_df['single'] == combination[1])] for combination in combinations}

pca_dim = 32

for combination, subset in zillow_subsets.items():
    # Sentence TF Embedding
    embeddings = np.array([model.encode(row['description']) for index, row in tqdm(subset.iterrows(), desc=f"Sentence TF Embedding for {combination}")])
    # Dimention Reduction
    pca = PCA(n_components=pca_dim)
    pca_transformed = pca.fit_transform(embeddings)
    # Store
    for i, index in enumerate(subset.index):
        embeddings_dict[subset.loc[index, 'zpid']] = embeddings[i] # original
        pca_embeddings_dict[subset.loc[index, 'zpid']] = pca_transformed[i] # pca
```

```python
embeddings_df = pd.DataFrame.from_dict(embeddings_dict, orient='index')
embeddings_df.columns = [f'emb_{i}' for i in range(embeddings_df.shape[1])]
embeddings_df.reset_index(inplace=True)
embeddings_df.rename(columns={'index': 'zpid'}, inplace=True)
embeddings_df.to_csv(os.path.join(main_dir, 'dataset', '5. stf_embedding.csv'), index=False)
embeddings_df
```

```python
embeddings_df = pd.DataFrame.from_dict(pca_embeddings_dict, orient='index')
embeddings_df.columns = [f'pca_{i}' for i in range(pca_dim)]
embeddings_df.reset_index(inplace=True)
embeddings_df.rename(columns={'index': 'zpid'}, inplace=True)
embeddings_df.to_csv(os.path.join(main_dir, 'dataset', '5. stf_pca.csv'), index=False)
embeddings_df
```

```python
import tiktoken

def num_tokens_from_string(string: str, encoding_name: str) -> int:
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```
